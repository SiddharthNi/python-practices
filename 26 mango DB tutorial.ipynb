Download MongoDB: Visit the official MongoDB website and download the appropriate version for your operating system.

Install MongoDB:

Follow the installation instructions provided on the MongoDB website for your specific operating system.

Start MongoDB: 

Once installed, start the MongoDB server. For most installations, you can start MongoDB by running the mongod command in your terminal or command prompt.

Basic Operations:

Connect to MongoDB: Open a new terminal or command prompt window and connect to the MongoDB server using the mongo command.

code

mongo

Create a Database: In MongoDB, you don't need to create a database explicitly. You can start using a database, and MongoDB will create it if it doesn't exist.

perl

code

use mydatabase

Create a Collection: Collections are similar to tables in relational databases. You can create a collection using the db.createCollection() method.

arduino

code

db.createCollection("mycollection")

Insert Documents: MongoDB stores data in JSON-like documents. You can insert documents into a collection using the insertOne() or insertMany() methods.

php

code

db.mycollection.insertOne({ name: "John", age: 30, city: "New York" })

Query Documents: You can query documents in MongoDB using the find() method.

lua
code

db.mycollection.find()

This will return all documents in the collection.

Update Documents: You can update documents using the updateOne() or updateMany() methods.

css

code

db.mycollection.updateOne({ name: "John" }, { $set: { age: 35 } })

Delete Documents: You can delete documents using the deleteOne() or deleteMany() methods.

css

code

db.mycollection.deleteOne({ name: "John" })
